```python
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

# Font sizes to use:
SMALLER_SIZE = 13
SMALL_SIZE = 14
MEDIUM_SIZE = 18
BIGGER_SIZE = 32

# General plot parameters:
plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=MEDIUM_SIZE)    # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALLER_SIZE)  # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
plt.rc('xtick.major', size=10)
plt.rc('xtick', direction="inout")
plt.rc('ytick.major', size=10)
plt.rc('ytick', direction="inout")

def minor(a, e): # Calculates semiminor axis
    return (math.sqrt(a**2 * (1 - e**2)))

# Main plotting function:
def orbit(shaded_belt, do_MMR, which_fig):
    plt.figure(figsize = (8,8), facecolor="white")
    ax = plt.gca()

    # Earth values:
    a = 1.0000010178  # semimajor axis [AU]
    e = 0.0167086     # eccentricity
    b = minor(a,e)    # semiminor axis [AU]
    
    # plotting variables
    linewid = 2.7       # linewidth variable for extra planet orbits
    linewid_r = 2       # linewidth variable for real planets
    alph = 1

    # shading the main Asteroid Belt region:
    if shaded_belt == True:
        ellipse = Ellipse(xy=(0,0), width=6.6, height=6.6, edgecolor=None, fc="gray", alpha=0.15,
                          label="AB", zorder= 0)
        ax.add_patch(ellipse)
        ellipse = Ellipse(xy=(0,0), width=4.2, height=4.2, edgecolor=None, fc="white", label="AB", zorder= 1)
        ax.add_patch(ellipse)

    # Earth:
    ellipse = Ellipse(xy=(0,0), width=(2*a), height=(2*b), edgecolor='black', fc='None',
                      lw=linewid_r, label="Earth", zorder= 20)
    ax.add_patch(ellipse)

    # Mars:
    ellipse = Ellipse(xy=(0,0), width=(2*1.524), height=(2*minor(1.524, 0.0934)), edgecolor="black",
                      fc='None', alpha = 0.4, lw=linewid_r, label="Mars", zorder= 20)
    ax.add_patch(ellipse)

    # Ceres:
    ellipse = Ellipse(xy=(0,0), width=(2*2.76618), height=(2*minor(2.76618, 0.07789)), edgecolor=c_c,
                      fc='None', alpha = alph, ls = "solid", lw=linewid, label="Ceres", zorder= 20)
    ax.add_patch(ellipse)
    
    # Pallas:
    ellipse = Ellipse(xy=(0,0), width=(2*2.77236), height=(2*minor(2.77236, 0.22965)), edgecolor=c_p,
                      fc='None', alpha = alph, ls = "solid", lw=linewid, label="Pallas", zorder= 20)
    ax.add_patch(ellipse)

    # Vesta:
    ellipse = Ellipse(xy=(0,0), width=(2*2.36076), height=(2*minor(2.36076, 0.0903)), edgecolor=c_v,
                      fc='None', alpha = alph, ls = "solid", lw=linewid, label="Vesta", zorder= 20)
    ax.add_patch(ellipse)

    # Hygeia:
    ellipse = Ellipse(xy=(0,0), width=(2*3.13687), height=(2*minor(3.13687, 0.11966)), edgecolor=c_h,
                      fc='None', alpha = alph, ls = "solid", lw=linewid, label="Hygeia", zorder= 20)
    ax.add_patch(ellipse)

    # Jupiter:
    ellipse = Ellipse(xy=(0,0), width=(2*5.2038), height=(2*minor(5.2038, 0.0489)), edgecolor="black",
                      fc='None', alpha = 0.4, lw=linewid_r, label="Jupiter", zorder= 2)
    ax.add_patch(ellipse)

    # Plotting MMR locations:
    linewid = 1.6       # using thinner line for MMRs
    if do_MMR == True:
        if which_fig == 'a':
            ellipse = Ellipse(xy=(0,0), width=(2*3.3), height=(2*3.3), edgecolor=c_e, fc='None', alpha = alph, ls = "dashed", lw=linewid, label="Hygeia", zorder= 20)
            ax.add_patch(ellipse)
            # 3J:2 at 3.972 AU [KANE]
            ellipse = Ellipse(xy=(0,0), width=(2*3.972), height=(2*3.972), edgecolor=c_e, fc='None', alpha = alph, ls = "dashed", lw=linewid, label="Hygeia", zorder= 20)
            ax.add_patch(ellipse)
            # 4J:1 at 2.065 AU [KANE]
            ellipse = Ellipse(xy=(0,0), width=(2*2.065), height=(2*2.065), edgecolor=c_e, fc='None', alpha = alph, ls = "dashed", lw=linewid, label="Hygeia", zorder= 20)
            ax.add_patch(ellipse)

            # 7J:4 at 3.58 AU [KANE]
            ellipse = Ellipse(xy=(0,0), width=(2*3.58), height=(2*3.58), edgecolor=c_e, fc='None', alpha = alph, ls = "dashed", lw=linewid, label="Hygeia", zorder= 20)
            ax.add_patch(ellipse)
        if which_fig == 'b':
            # 3J:1 at 2.5 AU ** [KANE]
            ellipse = Ellipse(xy=(0,0), width=(2*2.5), height=(2*2.5), edgecolor=c_e, fc='None', alpha = alph, ls = "dashed", lw=linewid, label="Hygeia", zorder= 20)
            ax.add_patch(ellipse)
            # 5J:2 at 2.82 AU ** 
            ellipse = Ellipse(xy=(0,0), width=(2*2.82), height=(2*2.82), edgecolor=c_e, fc='None', alpha = alph, ls = "dashed", lw=linewid, label="Hygeia", zorder= 20)
            ax.add_patch(ellipse)
            # 7J:3 at 2.96 AU
            ellipse = Ellipse(xy=(0,0), width=(2*2.96), height=(2*2.96), edgecolor=c_e, fc='None', alpha = alph, ls = "dashed", lw=linewid, label="Hygeia", zorder= 20)
            ax.add_patch(ellipse)
            # 2J:1 at 3.3 AU, commonly 3.3 AU ** [KANE]
            ellipse = Ellipse(xy=(0,0), width=(2*3.3), height=(2*3.3), edgecolor=c_e, fc='None', alpha = alph, ls = "dashed", lw=linewid, label="Hygeia", zorder= 20)
            ax.add_patch(ellipse)
            # 3J:2 at 3.972 AU [KANE]
            ellipse = Ellipse(xy=(0,0), width=(2*3.972), height=(2*3.972), edgecolor=c_e, fc='None', alpha = alph, ls = "dashed", lw=linewid, label="Hygeia", zorder= 20)
            ax.add_patch(ellipse)
            # 4J:1 at 2.065 AU [KANE]
            ellipse = Ellipse(xy=(0,0), width=(2*2.065), height=(2*2.065), edgecolor=c_e, fc='None', alpha = alph, ls = "dashed", lw=linewid, label="Hygeia", zorder= 20)
            ax.add_patch(ellipse)

# Colors associated w/ Ceres, Pallas, Vesta, Hygeia, and Earth (or asteroids):
c_c = 'xkcd:bright sky blue'
c_p = 'xkcd:bright blue'
c_v = 'green'
c_h = 'yellowgreen'
c_e = 'black'

# Choose Fig 2a or 2b:
which_fig = 'a'
orbit(True, True, which_fig)
plt.plot(0,0, color='black', marker = '*', markersize = 10) # Mark the Sun with a star

# Setting plot axis min/max:
if which_fig == 'a':
    plt.xlim(0,4.5)
    plt.ylim(0,4.5)
    plt.title("MMR locations")
elif which_fig == 'b':
    axmin = 2.1 - 0.3
    axmax = 3.3 + 0.1
    plt.xlim(axmin,axmax)
    plt.ylim(0,(axmax-axmin))
    plt.title("MMR locations, asteroid belt")

ax = plt.gca()
ax.set_aspect('equal', adjustable='box')

plt.xlabel("x (AU)")
plt.ylabel("y (AU)")

if which_fig == 'a': plt.savefig('fig_2a.png', dpi=300, bbox_inches="tight")
if which_fig == 'b': plt.savefig('fig_2b.png', dpi=300, bbox_inches="tight")

plt.show()
```
