```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Font sizes to use:
SMALLER_SIZE = 12
SMALL_SIZE = 16
MEDIUM_SIZE = 18
BIGGER_SIZE = 26

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)    # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
plt.rc('xtick.major', size=10)
plt.rc('xtick', direction="inout")
plt.rc('ytick.major', size=10)
plt.rc('ytick', direction="inout")
plt.rc('ytick.minor',size=5)

# For plotting all runs at once, had to take steps of 100 years due to hardware limitations;
# this doesn't change the visuals of the plot at all due to the large sample size.
# This is different than the numerical analysis, where it was possible to go through the
# runs one-by-one and therefore used every point (i.e., step size = 1 year).
step = 100

# ==========================================================================================
#   CERES
# ==========================================================================================
time = []
min_ceres_a = []
L_ceres_a = []
M_ceres_a = []
H_ceres_a = []
max_ceres_a = []
min_ceres_e = []
L_ceres_e = []
M_ceres_e = []
H_ceres_e = []
max_ceres_e = []
min_ceres_theta = []
L_ceres_theta = []
M_ceres_theta = []
H_ceres_theta = []
max_ceres_theta = []

with open("../bin/phaeton_min_ceres/data_in_mat/Earth", 'r') as f:  
    lines =  f.read().splitlines()[1::step]  
    for line in lines:
        time.append(float(line.split(None, 21)[0]))
        min_ceres_a.append(float(line.split(None, 21)[1]))
        min_ceres_e.append(float(line.split(None, 21)[2]))
        min_ceres_theta.append(float(line.split(None,21)[6]))

with open("../bin/phaeton_max_ceres/data_in_mat/Earth", 'r') as f:  
    lines =  f.read().splitlines()[1::step]
    for line in lines:
        max_ceres_a.append(float(line.split(None, 21)[1]))
        max_ceres_e.append(float(line.split(None, 21)[2]))
        max_ceres_theta.append(float(line.split(None,21)[6]))

with open("../bin/phaeton_L_ceres/data_in_mat/Earth", 'r') as f:  
    lines =  f.read().splitlines()[1::step]  
    for line in lines:
        L_ceres_a.append(float(line.split(None, 21)[1]))
        L_ceres_e.append(float(line.split(None, 21)[2]))
        L_ceres_theta.append(float(line.split(None,21)[6]))

with open("../bin/phaeton_M_ceres/data_in_mat/Earth", 'r') as f:  
    lines =  f.read().splitlines()[1::step]  
    for line in lines:
        M_ceres_a.append(float(line.split(None, 21)[1]))
        M_ceres_e.append(float(line.split(None, 21)[2]))
        M_ceres_theta.append(float(line.split(None,21)[6]))

with open("../bin/phaeton_H_ceres/data_in_mat/Earth", 'r') as f:  
    lines =  f.read().splitlines()[1::step]  
    for line in lines:
        H_ceres_a.append(float(line.split(None, 21)[1]))
        H_ceres_e.append(float(line.split(None, 21)[2]))
        H_ceres_theta.append(float(line.split(None,21)[6]))


# ==========================================================================================
#   PALLAS
# ==========================================================================================
min_pallas_a = []
L_pallas_a = []
M_pallas_a = []
H_pallas_a = []
max_pallas_a = []

min_pallas_e = []
L_pallas_e = []
M_pallas_e = []
H_pallas_e = []
max_pallas_e = []

min_pallas_theta = []
L_pallas_theta = []
M_pallas_theta = []
H_pallas_theta = []
max_pallas_theta = []

with open("../bin/phaeton_min_pallas/data_in_mat/Earth", 'r') as f:  
    lines =  f.read().splitlines()[1::step]  
    for line in lines:
        min_pallas_a.append(float(line.split(None, 21)[1]))
        min_pallas_e.append(float(line.split(None, 21)[2]))
        min_pallas_theta.append(float(line.split(None,21)[6]))

with open("../bin/phaeton_max_pallas/data_in_mat/Earth", 'r') as f:  
    lines =  f.read().splitlines()[1::step]
    for line in lines:
        max_pallas_a.append(float(line.split(None, 21)[1]))
        max_pallas_e.append(float(line.split(None, 21)[2]))
        max_pallas_theta.append(float(line.split(None,21)[6]))

with open("../bin/phaeton_L_pallas/data_in_mat/Earth", 'r') as f:  
    lines =  f.read().splitlines()[1::step]  
    for line in lines:
        L_pallas_a.append(float(line.split(None, 21)[1]))
        L_pallas_e.append(float(line.split(None, 21)[2]))
        L_pallas_theta.append(float(line.split(None,21)[6]))

with open("../bin/phaeton_M_pallas/data_in_mat/Earth", 'r') as f:  
    lines =  f.read().splitlines()[1::step]  
    for line in lines:
        M_pallas_a.append(float(line.split(None, 21)[1]))
        M_pallas_e.append(float(line.split(None, 21)[2]))
        M_pallas_theta.append(float(line.split(None,21)[6]))

with open("../bin/phaeton_H_pallas/data_in_mat/Earth", 'r') as f:  
    lines =  f.read().splitlines()[1::step]  
    for line in lines:
        H_pallas_a.append(float(line.split(None, 21)[1]))
        H_pallas_e.append(float(line.split(None, 21)[2]))
        H_pallas_theta.append(float(line.split(None,21)[6]))

# ==========================================================================================
#   VESTA
# ==========================================================================================
min_vesta_a = []
L_vesta_a = []
M_vesta_a = []
H_vesta_a = []
max_vesta_a = []

min_vesta_e = []
L_vesta_e = []
M_vesta_e = []
H_vesta_e = []
max_vesta_e = []

min_vesta_theta = []
L_vesta_theta = []
M_vesta_theta = []
H_vesta_theta = []
max_vesta_theta = []

with open("../bin/phaeton_min_vesta/data_in_mat/Earth", 'r') as f:  
    lines =  f.read().splitlines()[1::step]  
    for line in lines:
        min_vesta_a.append(float(line.split(None, 21)[1]))
        min_vesta_e.append(float(line.split(None, 21)[2]))
        min_vesta_theta.append(float(line.split(None,21)[6]))

with open("../bin/phaeton_max_vesta/data_in_mat/Earth", 'r') as f:  
    lines =  f.read().splitlines()[1::step]
    for line in lines:
        max_vesta_a.append(float(line.split(None, 21)[1]))
        max_vesta_e.append(float(line.split(None, 21)[2]))
        max_vesta_theta.append(float(line.split(None,21)[6]))

with open("../bin/phaeton_L_vesta/data_in_mat/Earth", 'r') as f:  
    lines =  f.read().splitlines()[1::step]  
    for line in lines:
        L_vesta_a.append(float(line.split(None, 21)[1]))
        L_vesta_e.append(float(line.split(None, 21)[2]))
        L_vesta_theta.append(float(line.split(None,21)[6]))

with open("../bin/phaeton_M_vesta/data_in_mat/Earth", 'r') as f:  
    lines =  f.read().splitlines()[1::step]  
    for line in lines:
        M_vesta_a.append(float(line.split(None, 21)[1]))
        M_vesta_e.append(float(line.split(None, 21)[2]))
        M_vesta_theta.append(float(line.split(None,21)[6]))

with open("../bin/phaeton_H_vesta/data_in_mat/Earth", 'r') as f:  
    lines =  f.read().splitlines()[1::step]  
    for line in lines:
        H_vesta_a.append(float(line.split(None, 21)[1]))
        H_vesta_e.append(float(line.split(None, 21)[2]))
        H_vesta_theta.append(float(line.split(None,21)[6]))

# ==========================================================================================
#   HYGEIA
# ==========================================================================================
min_hygeia_a = []
L_hygeia_a = []
M_hygeia_a = []
H_hygeia_a = []
max_hygeia_a = []

min_hygeia_e = []
L_hygeia_e = []
M_hygeia_e = []
H_hygeia_e = []
max_hygeia_e = []

min_hygeia_theta = []
L_hygeia_theta = []
M_hygeia_theta = []
H_hygeia_theta = []
max_hygeia_theta = []

with open("../bin/phaeton_min_hygeia/data_in_mat/Earth", 'r') as f:  
    lines =  f.read().splitlines()[1::step]  
    for line in lines:
        min_hygeia_a.append(float(line.split(None, 21)[1]))
        min_hygeia_e.append(float(line.split(None, 21)[2]))
        min_hygeia_theta.append(float(line.split(None,21)[6]))

with open("../bin/phaeton_max_hygeia/data_in_mat/Earth", 'r') as f:  
    lines =  f.read().splitlines()[1::step]
    for line in lines:
        max_hygeia_a.append(float(line.split(None, 21)[1]))
        max_hygeia_e.append(float(line.split(None, 21)[2]))
        max_hygeia_theta.append(float(line.split(None,21)[6]))

with open("../bin/phaeton_L_hygeia/data_in_mat/Earth", 'r') as f:  
    lines =  f.read().splitlines()[1::step]  
    for line in lines:
        L_hygeia_a.append(float(line.split(None, 21)[1]))
        L_hygeia_e.append(float(line.split(None, 21)[2]))
        L_hygeia_theta.append(float(line.split(None,21)[6]))

with open("../bin/phaeton_M_hygeia/data_in_mat/Earth", 'r') as f:  
    lines =  f.read().splitlines()[1::step]  
    for line in lines:
        M_hygeia_a.append(float(line.split(None, 21)[1]))
        M_hygeia_e.append(float(line.split(None, 21)[2]))
        M_hygeia_theta.append(float(line.split(None,21)[6]))

with open("../bin/phaeton_H_hygeia/data_in_mat/Earth", 'r') as f:  
    lines =  f.read().splitlines()[1::step]  
    for line in lines:
        H_hygeia_a.append(float(line.split(None, 21)[1]))
        H_hygeia_e.append(float(line.split(None, 21)[2]))
        H_hygeia_theta.append(float(line.split(None,21)[6]))

# ==========================================================================================
#   CONTROL
# ==========================================================================================
control_a = []
control_e = []
control_theta = []

with open("../bin/control_2M/data_in_mat/Earth", 'r') as f:
    lines =  f.read().splitlines()[1::step]
    for line in lines:
        control_a.append(float(line.split(None, 21)[1]))
        control_e.append(float(line.split(None, 21)[2]))
        control_theta.append(float(line.split(None,21)[6]))


# ==========================================================================================
#   DEFINING FUNCTIONS
# ==========================================================================================
# Defining functions for plotting:
def perihelion(a, e, i): # Calculate perihelion heliocentric distance [AU]
	return (a[i] * (1 - e[i]))

def aphelion(a, e, i): # Calculate aphelion heliocentric distance [AU]
	return (a[i] * (1 + e[i]))

def distance(rp, ra, a, e, theta, i): # Calculate Earth's heliocentric distance [AU]
	return ((rp[i] * ra[i])/(a[i] * (1 + (e[i] * math.cos(theta[i])))))

def calc_dist(mass):
	i = 0	# NB: i is increment variable for indexing, NOT inclination here
	while i < len(time):
		control_perihelion.append(perihelion(control_a, control_e, i))
		control_aphelion.append(aphelion(control_a, control_e, i))
		control_dist.append(distance(control_perihelion, control_aphelion, control_a, control_e, control_theta, i))

		if mass == "L":
			ceres_perihelion.append(perihelion(L_ceres_a, L_ceres_e, i))
			ceres_aphelion.append(aphelion(L_ceres_a, L_ceres_e, i))
			ceres_dist.append(distance(ceres_perihelion, ceres_aphelion, L_ceres_a, L_ceres_e, L_ceres_theta, i))
			pallas_perihelion.append(perihelion(L_pallas_a, L_pallas_e, i))
			pallas_aphelion.append(aphelion(L_pallas_a, L_pallas_e, i))
			pallas_dist.append(distance(pallas_perihelion, pallas_aphelion, L_pallas_a, L_pallas_e, L_pallas_theta, i))
			vesta_perihelion.append(perihelion(L_vesta_a, L_vesta_e, i))
			vesta_aphelion.append(aphelion(L_vesta_a, L_vesta_e, i))
			vesta_dist.append(distance(vesta_perihelion, vesta_aphelion, L_vesta_a, L_vesta_e, L_vesta_theta, i))
			hygeia_perihelion.append(perihelion(L_hygeia_a, L_hygeia_e, i))
			hygeia_aphelion.append(aphelion(L_hygeia_a, L_hygeia_e, i))
			hygeia_dist.append(distance(hygeia_perihelion, hygeia_aphelion, L_hygeia_a, L_hygeia_e, L_hygeia_theta, i))
		
		elif mass == "M":
			ceres_perihelion.append(perihelion(M_ceres_a, M_ceres_e, i))
			ceres_aphelion.append(aphelion(M_ceres_a, M_ceres_e, i))
			ceres_dist.append(distance(ceres_perihelion, ceres_aphelion, M_ceres_a, M_ceres_e, M_ceres_theta, i))
			pallas_perihelion.append(perihelion(M_pallas_a, M_pallas_e, i))
			pallas_aphelion.append(aphelion(M_pallas_a, M_pallas_e, i))
			pallas_dist.append(distance(pallas_perihelion, pallas_aphelion, M_pallas_a, M_pallas_e, M_pallas_theta, i))
			vesta_perihelion.append(perihelion(M_vesta_a, M_vesta_e, i))
			vesta_aphelion.append(aphelion(M_vesta_a, M_vesta_e, i))
			vesta_dist.append(distance(vesta_perihelion, vesta_aphelion, M_vesta_a, M_vesta_e, M_vesta_theta, i))
			hygeia_perihelion.append(perihelion(M_hygeia_a, M_hygeia_e, i))
			hygeia_aphelion.append(aphelion(M_hygeia_a, M_hygeia_e, i))
			hygeia_dist.append(distance(hygeia_perihelion, hygeia_aphelion, M_hygeia_a, M_hygeia_e, M_hygeia_theta, i))
		
		elif mass == "H":
			ceres_perihelion.append(perihelion(H_ceres_a, H_ceres_e, i))
			ceres_aphelion.append(aphelion(H_ceres_a, H_ceres_e, i))
			ceres_dist.append(distance(ceres_perihelion, ceres_aphelion, H_ceres_a, H_ceres_e, H_ceres_theta, i))
			pallas_perihelion.append(perihelion(H_pallas_a, H_pallas_e, i))
			pallas_aphelion.append(aphelion(H_pallas_a, H_pallas_e, i))
			pallas_dist.append(distance(pallas_perihelion, pallas_aphelion, H_pallas_a, H_pallas_e, H_pallas_theta, i))
			vesta_perihelion.append(perihelion(H_vesta_a, H_vesta_e, i))
			vesta_aphelion.append(aphelion(H_vesta_a, H_vesta_e, i))
			vesta_dist.append(distance(vesta_perihelion, vesta_aphelion, H_vesta_a, H_vesta_e, H_vesta_theta, i))
			hygeia_perihelion.append(perihelion(H_hygeia_a, H_hygeia_e, i))
			hygeia_aphelion.append(aphelion(H_hygeia_a, H_hygeia_e, i))
			hygeia_dist.append(distance(hygeia_perihelion, hygeia_aphelion, H_hygeia_a, H_hygeia_e, H_hygeia_theta, i))
			
		elif mass == "min":
			ceres_perihelion.append(perihelion(min_ceres_a, min_ceres_e, i))
			ceres_aphelion.append(aphelion(min_ceres_a, min_ceres_e, i))
			ceres_dist.append(distance(ceres_perihelion, ceres_aphelion, min_ceres_a, min_ceres_e, min_ceres_theta, i))
			pallas_perihelion.append(perihelion(min_pallas_a, min_pallas_e, i))
			pallas_aphelion.append(aphelion(min_pallas_a, min_pallas_e, i))
			pallas_dist.append(distance(pallas_perihelion, pallas_aphelion, min_pallas_a, min_pallas_e, min_pallas_theta, i))
			vesta_perihelion.append(perihelion(min_vesta_a, min_vesta_e, i))
			vesta_aphelion.append(aphelion(min_vesta_a, min_vesta_e, i))
			vesta_dist.append(distance(vesta_perihelion, vesta_aphelion, min_vesta_a, min_vesta_e, min_vesta_theta, i))
			hygeia_perihelion.append(perihelion(min_hygeia_a, min_hygeia_e, i))
			hygeia_aphelion.append(aphelion(min_hygeia_a, min_hygeia_e, i))
			hygeia_dist.append(distance(hygeia_perihelion, hygeia_aphelion, min_hygeia_a, min_hygeia_e, min_hygeia_theta, i))

		elif mass == "max":
			ceres_perihelion.append(perihelion(max_ceres_a, max_ceres_e, i))
			ceres_aphelion.append(aphelion(max_ceres_a, max_ceres_e, i))
			ceres_dist.append(distance(ceres_perihelion, ceres_aphelion, max_ceres_a, max_ceres_e, max_ceres_theta, i))
			pallas_perihelion.append(perihelion(max_pallas_a, max_pallas_e, i))
			pallas_aphelion.append(aphelion(max_pallas_a, max_pallas_e, i))
			pallas_dist.append(distance(pallas_perihelion, pallas_aphelion, max_pallas_a, max_pallas_e, max_pallas_theta, i))
			vesta_perihelion.append(perihelion(max_vesta_a, max_vesta_e, i))
			vesta_aphelion.append(aphelion(max_vesta_a, max_vesta_e, i))
			vesta_dist.append(distance(vesta_perihelion, vesta_aphelion, max_vesta_a, max_vesta_e, max_vesta_theta, i))
			hygeia_perihelion.append(perihelion(max_hygeia_a, max_hygeia_e, i))
			hygeia_aphelion.append(aphelion(max_hygeia_a, max_hygeia_e, i))
			hygeia_dist.append(distance(hygeia_perihelion, hygeia_aphelion, max_hygeia_a, max_hygeia_e, max_hygeia_theta, i))
		i += 1
```
