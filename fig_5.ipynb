```python
# Plot parameters:
SMALLER_SIZE = 12
SMALL_SIZE = 16
MEDIUM_SIZE = 18
BIGGER_SIZE = 26

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)     # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
plt.rc('xtick.major', size=10)
plt.rc('xtick', direction="inout")
plt.rc('ytick.major', size=10)
plt.rc('ytick', direction="inout")
plt.rc('ytick.minor',size=5)

# Colors associated w/ Ceres, Pallas, Vesta, Hygeia, and Earth:
c_c = 'xkcd:bright sky blue'
c_p = 'xkcd:bright blue'
c_v = 'green'
c_h = 'yellowgreen'
c_e = 'black'

# Setting up plot:
fig = plt.subplots(figsize =(8, 6), facecolor="white") 
plt.grid(axis="y", zorder=0)

labels = ['$0.01 M_$', '$1 M_$', '$2 M_$', '$5 M_$', '$10 M_$']
x = np.arange(len(labels))  # locations of mass labels
bar_width = 0.15            # set width of each bar

# Set position of bar on X axis 
br1 = np.arange(len(control_bar)) 
br2 = [x + bar_width for x in br1] 
br3 = [x + bar_width for x in br2] 
br4 = [x + bar_width for x in br3]
br5 = [x + bar_width for x in br4]
 
# Make the plot
plt.bar(br1, control_bar, color = c_e, width = bar_width, 
        edgecolor = 'black', label ='Reference solar system', zorder=10) 
plt.bar(br2, ceres_bar, color = c_c, width = bar_width, 
        edgecolor ='black', label ='Orb. Params. of Ceres', zorder=10) 
plt.bar(br3, pallas_bar, color = c_p, width = bar_width, 
        edgecolor ='black', label ='Orb. Params. of Pallas', zorder=10) 
plt.bar(br4, vesta_bar, color = c_v, width = bar_width, 
        edgecolor ='black', label ='Orb. Params. of Vesta', zorder=10) 
plt.bar(br5, hygeia_bar, color = c_h, width = bar_width, 
        edgecolor ='black', label ='Orb. Params. of Hygeia', zorder=10) 
 
# Adding x labels:
plt.xlabel('Mass group') 
plt.ylabel('Earth\'s Obliquity Variation $(\degree)$') 
plt.xticks([r + 2*bar_width for r in range(len(control_bar))], 
        ['$0.01 M_$', '$1 M_$', '$2 M_$', '$5 M_$', '$10 M_$'])

plt.legend()
plt.tight_layout()
plt.savefig('fig_5.png', dpi=300, bbox_inches="tight")

plt.show()
```
